```python

exportBR = dado[dado.type=="Export"]
importBR = dado[dado.type=="Import"]

exportBR_l5y = exportBR.loc['2014-01-01':'2019-01-01']
exportBR_l5y.product = exportBR_l5y.iloc[:,0]

fig,ax = plt.subplots()

ax.bar(['Oleo de soja','Farinha de soja','Soja'],
       [
         sum(exportBR_l5y[exportBR_l5y.product=='soybean_oil'].usd/1000000000),
            sum(exportBR_l5y[exportBR_l5y.product=='soybean_meal'].usd/1000000000),
               sum(exportBR_l5y[exportBR_l5y.product=='soybeans'].usd/1000000000)],
       color='green',width=0.3,label='Dolares')

ax.legend(loc='upper center',fontsize=16)
plt.title('Total de exportação em dolares para cada país no últimos 5 anos',fontsize=16)
plt.grid()
plt.ylabel('Bilhões de dolares',fontsize=16)

ax2=ax.twinx()

ax2.bar(['Oleo de soja','Farinha de soja','Soja'],
        [
         sum(exportBR_l5y[exportBR_l5y.product=='soybean_oil'].tons/1000000),
            sum(exportBR_l5y[exportBR_l5y.product=='soybean_meal'].tons/1000000),
               sum(exportBR_l5y[exportBR_l5y.product=='soybeans'].tons/1000000)],
        width=0.2,label='Toneladas',color='black')
ax2.set_ylabel('Milhões de toneladas',fontsize=16)
plt.legend(fontsize=16)
```
